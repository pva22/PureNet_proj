### Политика

```python
result = requests.get("https://news.mail.ru//politics/")
soup = BeautifulSoup(result.content, "lxml")
```

```python
d_pol = {}

for section in soup(class_="newsitem__title link-holder", href=True):
    href = section['href']
    for header in section.find_all(class_=['newsitem__title-inner']):
        title = header.string
    d_pol[title] = href
```

### Парсинг новости

```python
TAG_RE = re.compile(r'<[^>]+>')
def remove_tags(text):
    return TAG_RE.sub('', text)

def get_article(url):
    
    result = requests.get("https://news.mail.ru/" + url)
    soup = BeautifulSoup(result.content, "lxml")

    a = ''
    m = ''
    q = ''
    for section in soup(class_=["article__intro",
                         "article__item article__item_alignment_left article__item_html",
                         "article-keyphrase__inner"]):
        if section.p != None:
            a += (' ' + remove_tags(str(section.p)))
    #shor main
    for section in soup(class_=["article-keyphrase__inner"]):
        if section.string != None:
            m += (' ' + remove_tags(str(section.string)))
    #quote       
    for section in soup(class_=["quote__text", "author__name","author__text quote__position"]):
        if section.string != None:
            q += (' ' + remove_tags(str(section.string)))
    
    a = a.replace(u'\xa0', u' ')
    m = m.replace(u'\xa0', u' ')
    q = q.replace(u'\xa0', u' ')
    return a,m,q
```

```python
art = get_article('/politics/34893427/')
```
